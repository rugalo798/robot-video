```
import pandas as pd
import bar_chart_race as bcr
```

```
df_covid = pd.read_excel('./HIST_PAINEL_COVIDBR_17jun2020.xlsx')
```

```
df_covid.head()
```

```
df_covid.dropna(subset=["municipio"], inplace=True)
```

```
df_covid.head()
```

```
df_covid_bcr = pd.DataFrame(df_covid.groupby(["regiao", "data"]).sum()['casosAcumulado']).unstack().T.droplevel(level=0)
```

```
df_covid_bcr
```

```
bcr.bar_chart_race(df_covid_bcr)#Criar corrida de graficos por regiões
```

```
df_covid_bcr_estados = pd.DataFrame(df_covid.groupby(["estado", "data"]).sum()['casosAcumulado']).unstack().T.droplevel(level=0)#Criar um data frame para os estados
```

```
bcr.bar_chart_race(df_covid_bcr_estados)#Criar uma corrida de graficos por estados
```
